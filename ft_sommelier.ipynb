# V.1: Exploring the green reds

### a) Write a function that will plot a scatterplot matrix of your red wine data.

```python
def plot_scatter_matrix(wine_data, good_threshold, bad_threshold, save_plot=False):
    numvars, numdata = wine.shape
    fig, axes = plt.subplots(numdata, numdata, figsize = (100, 100))
    fig.subplots_adjust(wspace = 0, hspace = 0)
    g = wine.quality > good_threshold
    b = wine.quality < bad_threshold
    g_i = wine[g].index
    b_i = wine[b].index
    for  i in range(numdata):
        for j in range(numdata):
            if (i == j):
                axes[j, i].text(0.5, 0.5, wine.columns[i], fontsize = 40, ha = 'center')
            else :
                plt.subplot(numdata, numdata, i * numdata + j + 1)
                x1 = wine.iloc[b_i, j]
                y1 = wine.iloc[b_i, i]

                x2 = wine.iloc[g_i, j]
                y2 = wine.iloc[g_i, i]
                plt.plot(x1, y1, 'ro', x2, y2, 'go', markersize = 10)
        j = 0
    if(save_plot):
        fig.savefig('scatterplot_matrix.png')
    plt.show()
    
```
